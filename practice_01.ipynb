```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has a build in dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

visiualize the data

```python
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',size='size')
```

house price dataset

```python
from sklearn.datasets import load_diabetes
diabetis = load_diabetes()
df=pd.DataFrame(diabetis.data,columns=diabetis.feature_names)
df['age']=diabetis.target
df.head()
sns.displot(df['age'])
```

correlation 

```python
correlation=df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
